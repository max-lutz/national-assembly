## Extract all the deputies infos from the csv files

The next cell scans through all the files with deputies info (they start with 'PA') and extracts the relevant data (name, family name, date of birth, political party...)
At the end we output a list of political parties with their respective code.

```python
k = 0
j = 0
w = 0

x = set()
notebook_path = os.path.dirname(os.path.abspath("Data exploratory phase.ipynb"))
for path, subdirs, files in os.walk(notebook_path + '\json'):
    for name in files:
        #print(os.path.join(path, name))
        if name[0:2] == 'VT':
            with open(os.path.join(path, name)) as f:
                data = json.load(f)
                print(data['scrutin']['dateScrutin'])
                print(data['scrutin']['typeVote']['libelleTypeVote'])
                print(data['scrutin']['titre'])
                print(data['scrutin']['demandeur']['texte'])
                print(data['scrutin']['syntheseVote']['nombreVotants'])
                print(data['scrutin']['syntheseVote']['nbrSuffragesRequis'])
                print(data['scrutin']['syntheseVote']['decompte']['nonVotants'])
                print(data['scrutin']['syntheseVote']['decompte']['pour'])
                print(data['scrutin']['syntheseVote']['decompte']['contre'])
                print(data['scrutin']['syntheseVote']['decompte']['abstentions'])
                print(data['scrutin']['syntheseVote']['decompte']['nonVotantsVolontaires'])
```

```python
column_names = ["code", "date", "type", "titre", "demandeur", "nb votants", "requis", "non votants", "pour", "contre", "abstentions", "non votants volontaires"]
df_vote = pd.DataFrame(columns = column_names)
row_index = 0
for path, subdirs, files in os.walk(notebook_path + '\json'):
    for name in files:
        #print(os.path.join(path, name))
        if name[0:2] == 'VT':
            with open(os.path.join(path, name)) as f:
                data = json.load(f)
                l = [data['scrutin']['uid']]
                l.append(data['scrutin']['dateScrutin'])
                l.append(data['scrutin']['typeVote']['libelleTypeVote'])
                l.append(data['scrutin']['titre'])
                l.append(data['scrutin']['demandeur']['texte'])
                l.append(data['scrutin']['syntheseVote']['nombreVotants'])
                l.append(data['scrutin']['syntheseVote']['nbrSuffragesRequis'])
                l.append(data['scrutin']['syntheseVote']['decompte']['nonVotants'])
                l.append(data['scrutin']['syntheseVote']['decompte']['pour'])
                l.append(data['scrutin']['syntheseVote']['decompte']['contre'])
                l.append(data['scrutin']['syntheseVote']['decompte']['abstentions'])
                l.append(data['scrutin']['syntheseVote']['decompte']['nonVotantsVolontaires'])
                df_vote.loc[row_index] = l
                print(row_index)
                row_index = row_index + 1
df_vote.info()
```
