## Extract all the deputies infos from the csv files

The next cell scans through all the files with deputies info (they start with 'PA') and extracts the relevant data (name, family name, date of birth, political party...)
At the end we output a list of political parties with their respective code.

```python
k = 0
j = 0
w = 0

x = set()
notebook_path = os.path.dirname(os.path.abspath("Data exploratory phase.ipynb"))
for path, subdirs, files in os.walk(notebook_path + '\json'):
    for name in files:
        #print(os.path.join(path, name))
        if name[0:2] == 'PA':
            with open(os.path.join(path, name)) as f:
                data = json.load(f)
                print(data['acteur']['etatCivil']['ident']['civ'], data['acteur']['etatCivil']['ident']['nom'], data['acteur']['etatCivil']['ident']['prenom'], '   ', data['acteur']['etatCivil']['infoNaissance']['dateNais'], '   ', data['acteur']['profession']['socProcINSEE']['famSocPro'], '   ',data['acteur']['uid']['#text'])
                #print(data['acteur']['etatCivil']['ident']['prenom'])
                #print(data['acteur']['etatCivil']['ident']['nom'])
                #print(data['acteur']['etatCivil']['infoNaissance']['dateNais'])
                if data['acteur']['etatCivil']['ident']['civ'] == 'Mme':
                    w = w + 1
                for i in range(len(data['acteur']['mandats']['mandat'])):
                    if data['acteur']['mandats']['mandat'][i]['typeOrgane'] == 'PARPOL':
                        #print(data['acteur']['mandats']['mandat'][i]['typeOrgane'])
                        #print(data['acteur']['mandats']['mandat'][i]['organes']['organeRef'])
                        x.add(data['acteur']['mandats']['mandat'][i]['organes']['organeRef'])
                    if data['acteur']['mandats']['mandat'][i]['typeOrgane'] == 'PARPOL':
                        k = k + 1
                j = j + 1
print('Number of deputies affiliated to a political party : ',k,'/',j)
print('Number of women deputies : ',w,'/',j)
print('Political parties : ', x)
```

## Extract the political parties infos
First we get the number of members per political parties

```python
dic = {"PO744856" : 0, 
       "PO710611" : 0, 
       "PO684936" : 0, 
       "PO685809" : 0, 
       "PO684926" : 0, 
       "PO761294" : 0, 
       "PO761239" : 0, 
       "PO684932" : 0, 
       "PO744860" : 0, 
       "PO705103" : 0, 
       "PO684938" : 0, 
       "PO685893" : 0, 
       "PO685813" : 0, 
       "PO744864" : 0, 
       "PO746314" : 0, 
       "PO744862" : 0, 
       "PO684934" : 0, 
       "PO710396" : 0, 
       "PO744858" : 0
      }

#count the number of deputies per political parties
for path, subdirs, files in os.walk(notebook_path + '\json'):
    for name in files:
        #print(os.path.join(path, name))
        if name[0:2] == 'PA':
            with open(os.path.join(path, name)) as f:
                data = json.load(f)
                for i in range(len(data['acteur']['mandats']['mandat'])):
                    if data['acteur']['mandats']['mandat'][i]['typeOrgane'] == 'PARPOL':
                        dic[data['acteur']['mandats']['mandat'][i]['organes']['organeRef']] += 1

print(dic)
```

Let's save all this in a dataframe

```python
column_names = ["code", "name", "abreviated_name", "members"]
df_polpar = pd.DataFrame(columns = column_names)
row_index = 0

for path, subdirs, files in os.walk(notebook_path + '\json'):
    for name in files:
        #print(os.path.join(path, name))
        if name[0:2] == 'PO':
            with open(os.path.join(path, name)) as f:
                data = json.load(f)
                if data['organe']['codeType'] == 'PARPOL':
                    l = [data['organe']['uid']]
                    l.append(data['organe']['libelle'])
                    l.append(data['organe']['libelleAbrev'])
                    l.append(dic[data['organe']['uid']])
                    print(l)
                    df_polpar.loc[row_index] = l
                    row_index = row_index + 1 
df_polpar.info()
```
