```python
import pandas as pd
import numpy as np
import json
import os
import requests, zipfile, io
import shutil
from PIL import Image
from os import listdir
from os.path import isfile, join
```

```python
df_deputies = pd.read_csv(os.path.join(os.getcwd(), 'data', 'df_dep.csv'))
deputies_list = df_deputies['code'].to_list()
```

```python
i = 0
for deputy_code in deputies_list:
    i = i + 1
    link = 'https://datan.fr/assets/imgs/deputes_nobg_webp/depute_' + deputy_code[2:] + '_webp.webp'
    r = requests.get(link)
    file = open(os.path.join(os.getcwd(), 'data', 'pictures')+'\\'+deputy_code+'.webp', 'wb')
    file.write(r.content)
    percentage = round(i/573*100,2)
    print('Downloading image at ' + link + '   ' + str(percentage) + '%')
    file.close()
```

```python
from webptools import dwebp
```

```python
i = 0
mypath = os.path.join(os.getcwd(), 'data', 'pictures')
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for file in onlyfiles:
    i = i + 1
    percentage = round(i/573*100,2)
    path = os.path.join(os.getcwd(), 'data', 'pictures', file)
    dwebp(path, path[:-4]+'jpg',"-o")
    print(str(percentage) + '%')
    os.remove(path)
```
